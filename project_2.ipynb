```python
import numpy as np
import scipy.integrate
import matplotlib.pyplot as plt

def bessel_j(x, n):
    def integrand(theta):
        if n == 0:
            return np.cos(x * np.sin(theta))
        else:
            return np.cos(x * np.sin(theta) - n * theta)
    val, err = scipy.integrate.quad(integrand, 0, np.pi)
    return (1/np.pi) * val

x = np.linspace(0, 20, 100)

# Calculate Bessel functions for n=0,1,2
y0 = [bessel_j(i, 0) for i in x]
y1 = [bessel_j(i, 1) for i in x]
y2 = [bessel_j(i, 2) for i in x]


plt.figure(figsize=(10, 6))
plt.plot(x, y0, label='n=0')
plt.plot(x, y1, label='n=1')
plt.plot(x, y2, label='n=2')
plt.xlabel('x')
plt.ylabel('J_n(x)')
plt.title('Bessel Functions of Orders 0, 1, 2')
plt.grid(True, alpha=0.7)
plt.legend()

plt.show()
```
